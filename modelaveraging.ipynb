```python
%matplotlib inline
import matplotlib.style
matplotlib.style.use('ggplot')

import matplotlib.pyplot as plt
import sympy
import scipy.stats
scipy.stats.norm

from sympy import init_printing, pprint, oo, lambdify 
from sympy.stats import Normal, density, E, variance
init_printing(use_latex=True)
import numpy as np

#Aan de andere kant, tot 2030 is de overlap in de banden
#zodanig groot, dat ik gewoon low-low tot high-high als uitgangspunt zou
#nemen voor een 5-95 range voor trends uit de scenarios. 
#year,low_05,low_95,high_05,high_95,comment
#2015,2.3,5.2,3.9,6.7
```

```python
ll = 2.3
ul = 6.7
mean_knmi =  (ll+ul)/2
sigma_times_z90 = ul - (ll+ul)/2
sigma_times_z90
N = Normal('standard', 0, 1)
N = scipy.stats.norm(scale=1, loc=0)
sigma_knmi = sigma_times_z90/N.ppf(0.95)
N_knmi =  scipy.stats.norm(scale=sigma_knmi, loc=mean_knmi)
N_knmi.ppf(0.05), N_knmi.ppf(0.95), N_knmi.cdf(2.3), N_knmi.cdf(6.7)

sigma_hkv = 0.5
mean_hkv = 2
N_hkv = scipy.stats.norm(scale=sigma_hkv, loc=mean_hkv)
```

```python
x = np.linspace(0,10, num=100)
plt.fill_between(x, N_hkv.pdf(x), facecolor='blue', alpha=0.5)
plt.fill_between(x, N_knmi.pdf(x), facecolor='black', alpha=0.5)
plt.fill_between(x, (N_hkv.pdf(x) + N_knmi.pdf(x))/2.0, facecolor='green', alpha=0.5)
```

```python
a, b, x, y = sympy.symbols('a b x y')
sigma1 = sympy.Symbol('sigma1', positive=True)
sigma2 = sympy.Symbol('sigma2', positive=True)
mu1 = sympy.Symbol('mu1')
mu2 = sympy.Symbol('mu2')

```

```python
N1 = Normal("model A", mu1, sigma1)
pdf1 = density(N1)(x)
N2 = Normal("model B", mu2, sigma2)
pdf2 = density(N2)(x)
```

```python
from IPython.html.widgets import interact
from IPython.display import display
def formula(mu1_, sigma1_):
    pdf = pdf1.subs({mu1: mu1_, sigma1: sigma1_})/2 + pdf2.subs({mu2:0, sigma2: 1})/2
    display(pdf)
    cdf = pdf.integrate((x, -oo, oo)).simplify()
    display(cdf)
interact(formula, mu1_=(0,2), sigma1_=(1,3))
```

```python

pdf = pdf1.subs({mu1: 0, sigma1: 1})/2 + pdf2.subs({mu2:2, sigma2: 0.5})/2
pdf.integrate((x, -oo, oo)).simplify()
X = np.linspace(-3,15, num=100)
d = lambdify(x, pdf, np)(X)
plt.title('$\hat{x}$=%.1f' % (np.sum(X*d)/np.sum(d)).item())
plt.plot(X,d)
```
